```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 20 # number of points
a = 1 # upper bound
b = 0 # lower bound

# Random coordinates [b,a) uniform distributed
coordy = (b - a) *  np.random.random_sample((n,)) + a # generate random y
coordx = (b - a) *  np.random.random_sample((n,)) + a # generate random x

# Create limits (x,y)=((-5,5),(-5,5))
plt.xlim((b,a))
plt.ylim((b,a))
X_train=[]
y_train=[]
# Plot points
for i in range(n):
    plt.plot(coordx[i],coordy[i],'ro')
    X_train.append([coordx[i],coordy[i]])
    if (coordx[i]>coordy[i]):
      y_train.append(1)
    else:
      y_train.append(-1)
plt.show()
print(X_train)
print(y_train)



```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 1000 # number of points
a = 1 # upper bound
b = 0 # lower bound

# Random coordinates [b,a) uniform distributed
coordy = (b - a) *  np.random.random_sample((n,)) + a # generate random y
coordx = (b - a) *  np.random.random_sample((n,)) + a # generate random x

# Create limits (x,y)=((-5,5),(-5,5))
plt.xlim((b,a))
plt.ylim((b,a))
X_test=[]
y_test=[]
# Plot points
for i in range(n):
    plt.plot(coordx[i],coordy[i],'ro')
    X_test.append([coordx[i],coordy[i]])
    if (coordx[i]>coordy[i]):
      y_test.append(1)
    else:
      y_test.append(-1)
plt.show()
print(X_test)
print(y_test)

```

```python
from sklearn.linear_model import Perceptron, SGDClassifier
```

perceptron

```python
clf = Perceptron(tol=1e-3, random_state=0)
clf.fit(X_train, y_train)

clf.score(X_test, y_test)
```

after change train set

```python
clf = Perceptron(tol=1e-3, random_state=0)
clf.fit(X_train, y_train)

clf.score(X_test, y_test)
```

hinge

```python
clf = SGDClassifier(loss='hinge',max_iter=1000, tol=1e-3)
clf.fit(X_train, y_train)


clf.score(X_test, y_test)
```
